```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import display
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score, train_test_split

path = "/Users/alexis/Documents/Scolarité/REPO/projet-ml/data/spambase"
```

```python
# faire une liste names avec le nom des colonnes ? --> dans le pre-processing sur git oui
spam_df = pd.read_csv("/Users/alexis/Documents/Scolarité/REPO/projet-ml/data/spambase/spambase.data", header=None)
```

On observe qu'il y a 40% de mails spams. Donc on a un dataset légèrement déséquilibré.

```python
corr = spam_df.corr()
print(corr)
sns.heatmap(corr, cmap="coolwarm")
```

```python
# On réduit le dataset pour observer les features de 24 à 40 qui sont plus corrélées que les autres
spam_df_reduced = spam_df.iloc[:, 24:40]
display(spam_df_reduced.corr())
sns.heatmap(spam_df_reduced.corr(), cmap="coolwarm")
```

```python
# on souhaite désormais séparer les données en entrainement et test
# est-ce que je rajoute validation ? je ne pense pas car dataset petit. On choisit 0.2 de test car petit dataset.
# random_state=0 pour rendre l'expérience reproductible
X_train, X_test, y_train, y_test = train_test_split(
    spam_df.iloc[:, :57], spam_df.iloc[:, 57], test_size=0.3, shuffle=True, random_state=0
)


print(y_test[:2])
y_test.shape
X_train.shape
type(X_train)
```

```python
# on applique la Random Forest
forest = RandomForestClassifier(n_estimators=50, random_state=0)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)
print(f"The accuracy for the training set is {forest.score(X_train, y_train):.4f}")
print(f"The accuracy for the test set is {forest.score(X_test, y_test):.4f}")
print(classification_report(y_test, y_pred))
# Faire une comparaison entre TP et FN
# pour la présentation des résultats : mettre spam 0 / spam 1
```

Pour la classe spam, on a un recall de 90%. Cela signifie que le modèle a identifié 90% des mails spams. On peut considérer ça comme un bon score.   
La précision pour la classe spam = 96%. Cela signifie que 96% des mails classés spams sont bien spams.  
Accuracy globale de 95% : 95% des mails sont bien classés. 
Les deux F1-score sont très bons : il y a un bon équilibre entre précision et recall pour la classe spam et non-spam.  
On effectue une cross-val pour vérifier que le modèle généralise bien. 

```python
# on évalue le score de cross-val avec une validation à 5 plis
scores = cross_val_score(forest, spam_df.iloc[:, :57], spam_df.iloc[:, 57], cv=5)
mean_score = np.mean(scores)
print(f"Cross validation score is {mean_score}")
```

```python
scores = cross_val_score(forest, spam_df.iloc[:, :57], spam_df.iloc[:, 57], cv=5)
mean_scores = np.mean(scores)
print(f"The cross-validation scores for 5 folds is {mean_scores: .3f}")
```

Puisque le score est de 93%, on peut considérer qu'il y a une bonne généralisation. 

```python
# print(len(forest.feature_importances_))
plt.barh(range(57), forest.feature_importances_, align="center")
plt.xlabel("Feature importance")
plt.ylabel("Feature")
```

Aucune importance ne dépasse 0.1 : on peut considérer que le modèle n'est dépendant d'aucune feature en particulier. Chaque feature apporte son information.
